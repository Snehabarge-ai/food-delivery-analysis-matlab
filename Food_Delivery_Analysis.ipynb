```python
import pandas as pd
import json
import sqlite3

# Load orders
orders = pd.read_csv("orders.csv")

# Load users
with open("users.json") as f:
    users = pd.DataFrame(json.load(f))

# Load restaurants from SQL (if DB)
conn = sqlite3.connect("restaurants.db")
restaurants = pd.read_sql("SELECT * FROM restaurants", conn)

# Merge datasets
data = orders.merge(users, on="user_id", how="left")
final_data = data.merge(restaurants, on="restaurant_id", how="left")

# Save final dataset
final_data.to_csv("final_food_delivery_dataset.csv", index=False)

final_data.head()
```
